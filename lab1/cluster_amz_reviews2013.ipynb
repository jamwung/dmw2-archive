```python
def merge_review_parts(df):
    """Merge the 'review_headline' and the 'review_body' columns of `df`."""
    merged_text = df.review_headline + ' ' + df.review_body
    return merged_text

def clean_text(text, stop_words, lemmatizer):
    """Preprocess the text using lemmatization."""
    text = text.casefold()
    text_list = [
        lemmatizer.lemmatize(word)
        for word in re.findall(r'\b[a-z-]+\b', text)
        if word not in stop_words
    ]
    return ' '.join(text_list)

def lemmatize(df, stop_words, lemmatizer):
    """Lemmatize the text."""
    lemmd_text = (merge_review_parts(df)
                  .apply(lambda x: clean_text(x, stop_words, lemmatizer))
                 )
    return lemmd_text

def vectorize(corpus, model, params):
    """
    Vectorize the corpus of articles using `model`.
    """
    vectorizer = model(**params)
    sparse_corpus = vectorizer.fit_transform(corpus)
    corpus_df = pd.DataFrame.sparse.from_spmatrix(
        sparse_corpus, columns=vectorizer.get_feature_names_out()
    )
    return corpus_df

def save_pkl(obj, name, prompt=True):
    """Save an object to a pickle file."""
    folder = 'pickles'
    ext = '.pkl'
    if not os.path.exists(folder):
        os.makedirs(folder)
    if name[-4:] == ext:
        fp = os.path.join(folder, name)
    else:
        fp = os.path.join(folder, name+ext)
    joblib.dump(obj, fp)
    if prompt:
        print('Object pickled for future use.')
    return

def load_pkl(obj, name):
    """Load an object from a pickle file."""
    folder = 'pickles'
    ext = '.pkl'
    if not os.path.exists(folder):
        raise ValueError("'pickles' folder does not exist.")
    if name[-4:] == ext:
        fp = os.path.join(folder, name)
    else:
        fp = os.path.join(folder, name+ext)
    return joblib.load(fp)
```

```python
try:
    models_reviews_2013 = load_pkl(models_reviews_2013, 'models_reviews_2013')
    scores_reviews_2013 = load_pkl(scores_reviews_2013, 'scores_reviews_2013')
except:
    models_reviews_2013 = {}
    scores_reviews_2013 = {'sse': [], 'ch': [], 'sh': [], 'db': []}
    for k in tqdm(range(2, 11)):
        kmeans = KMeans(n_clusters=k, random_state=1337)
        cluster_labels = kmeans.fit_predict(docterm_matrix_2013)
        models_reviews_2013.update({k: {
            'model': kmeans,
            'labels': cluster_labels,
        }})
        scores_reviews_2013['sse'].append(kmeans.inertia_)
        scores_reviews_2013['ch'].append(calinski_harabasz_score(
            docterm_matrix_2013.to_numpy(),
            cluster_labels
        ))
        scores_reviews_2013['sh'].append(silhouette_score(
            docterm_matrix_2013.to_numpy(), cluster_labels
        ))
        scores_reviews_2013['db'].append(davies_bouldin_score(
            docterm_matrix_2013.to_numpy(), cluster_labels
        ))

    save_pkl(models_reviews_2013, 'models_reviews_2013')
    save_pkl(scores_reviews_2013, 'scores_reviews_2013')
```
